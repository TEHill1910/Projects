```python
import fastf1
fastf1.Cache.enable_cache("cache")

import pandas as pd
```

```python
def get_race_results(year, gp_name):
    """
    Loads race results + pit stop count + tyre strategy.
    Compatible with FastF1 versions WITHOUT get_pitstops().
    """
    try:
        session = fastf1.get_session(year, gp_name, "R")
        session.load()

        results = session.results.copy()
        laps = session.laps.copy()

        # --------------------------------------------------------
        # PIT STOP COUNT (works for ALL FastF1 versions)
        # --------------------------------------------------------
        pit_laps = laps[
            laps["PitInTime"].notna() | laps["PitOutTime"].notna()
        ].copy()

        pit_counts = (
            pit_laps.groupby("Driver")["Driver"]
            .count()
            .rename("PitStops")
        )

        # --------------------------------------------------------
        # TYRE STRATEGY (SOFT -> MEDIUM -> HARD)
        # --------------------------------------------------------
        stint_info = (
            laps[["Driver", "Stint", "Compound"]]
            .dropna()
            .sort_values(["Driver", "Stint"])
            .drop_duplicates(subset=["Driver", "Stint"])  # get first compound of each stint
        )

        strategy = (
            stint_info.groupby("Driver")["Compound"]
            .apply(lambda x: list(x))
            .rename("TyreStrategy")
        )

        # --------------------------------------------------------
        # MERGE EVERYTHING
        # --------------------------------------------------------
        df = results.merge(
            pit_counts,
            left_on="Abbreviation",
            right_index=True,
            how="left"
        )

        df = df.merge(
            strategy,
            left_on="Abbreviation",
            right_index=True,
            how="left"
        )

        df["PitStops"].fillna(0, inplace=True)

        df["GP"] = gp_name
        df["Year"] = year

        return df

    except Exception as e:
        print(f"Error loading race results for {gp_name} {year}: {e}")
        return None
```

```python
def get_race_list(year):
    schedule = fastf1.get_event_schedule(year, include_testing=False)


    return list(schedule["EventName"])
```

```python
all_results = []
years = [2022,2023,2024,2025]

for year in years:
    races = get_race_list(year)

    for gp in races:
        print(f"Processing race results for {gp} {year}")
        df = get_race_results(year, gp)
        if df is not None:
            all_results.append(df)
```

```python

results_df = pd.concat(all_results, ignore_index=True)



display(results_df.columns)


```

```python


columns = [
    'DriverNumber',
    'Abbreviation',
    'GridPosition',
    'ClassifiedPosition',
    'TyreStrategy',
    'Status',
    'GP',
    'Year'
]


results_df = results_df[columns]

results_df = results_df.rename(columns={"Abbreviation":"Driver"})

results_df
```

```python
results_df.to_csv("f1_race_results.csv", index=False)
```
